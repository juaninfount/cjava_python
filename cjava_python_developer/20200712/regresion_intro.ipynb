```python
import numpy as np
from sklearn import datasets, linear_model
import matplotlib.pyplot as plt
```

```python
# imprimir datasets
boston = datasets.load_boston()
```

```python
#entendimiento de la data
# 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM'
print(boston.keys())
# data, obejtivos, nombres de caracteristicas, descripcion, nombre de archivo
```

```python
print(boston.data.shape) # N° filas, N° columnas
```

```python
# Preparar la data de la regresion polinomial
X_p = boston.data[:,np.newaxis,5] # columna 5, numero de hab
```

```python
Y_p = boston.target # media de las casas de boston
len(Y_p)
#len(X_p)
```

```python
plt.scatter(X_p,Y_p)
plt.show()
```

```python
# Implementacion de la regresion polinomial
from sklearn.model_selection import train_test_split
```

```python
X_train_p, X_test_p, y_train_p, y_test_p = train_test_split( X_p, Y_p, test_size=0.2) 
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poli_reg = PolynomialFeatures(degree=2) # cuadratico
X_train_poli = poli_reg.fit_transform(X_train_p)
X_test_poli = poli_reg.fit_transform(X_test_p)
pr = linear_model.LinearRegression()
```

```python
pr.fit(X_train_poli, y_train_p) # entrenar modelo
```

```python
Y_pred_pr = pr.predict(X_test_poli)
plt.scatter(X_test_p, y_test_p) # diagrama de puntos o dispersion
plt.plot(X_test_p, Y_pred_pr,color='red',linewidth=3)
plt.show()
```
