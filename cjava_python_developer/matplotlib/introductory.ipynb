```python
# Object oriented style 
import matplotlib.pyplot as plt

x = np.linspace(0,2,100) # array de 100 elementos, desde 0 hasta 2 en una tasa uniforme
fig, ax = plt.subplots() # crear una figura y los ejes
ax.plot(x,x,label='linear') # plotear x data en los ejes
ax.plot(x,x**2,label='quadractic') # plotear x**2  data en los ejes
ax.plot(x,x**3,label='cubic') # plotear x**3  data en los ejes
ax.set_xlabel('x label') 
ax.set_ylabel('y label')
ax.set_title('Simple plot') # title
ax.legend() # leyenda
```

```python
# pyplot style
x = np.linspace(0,2,100) # array de 100 elementos, desde 0 hasta 2 en una tasa uniforme

plt.plot(x,x,label='linear') # plotear x data en los ejes
plt.plot(x,x**2,label='quadractic') # plotear x**2  data en los ejes
plt.plot(x,x**3,label='cubic') # plotear x**3  data en los ejes
plt.xlabel('x label') 
plt.ylabel('y label')
plt.title('Simple plot') # title
plt.legend() # leyenda
# import pandas as pd
```
