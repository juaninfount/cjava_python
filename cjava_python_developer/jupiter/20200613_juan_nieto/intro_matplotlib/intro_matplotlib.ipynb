$$f(x) = x^2 - y^2$$

```python
def f(x,y):
    return x**2 - y**2
```

```python
x = np.linspace(-2,2)
y = np.linspace(-2,2)
xx, yy =np.meshgrid(x,y)
zz = f(xx,yy)
plt.contourf(xx,yy,zz) # la f es de fill colour
plt.colorbar()
```

```python
x = np.linspace(-1,7,1000)  
fig = plt.figure()  
plt.subplot(211) # Add a subplot to the current figure. position 2,1,1
plt.plot(x,np.sin(x))  
plt.grid(False) # remover grid
plt.title('Funcion seno')
fig.savefig('grafica_seno.png')

plt.subplot(2,1,2)
plt.plot(x, np.cos(x))
plt.grid(False)
plt.title('Funcion coseno')
fig.savefig('grafica_coseno.png')
```

**Ejercicio**
Crear una función que represente gráficamente esta expresión:
$$\sin(2 \pi f_1 t) + \sin(2 \pi f_2 t)$$
Siendo $f_1$ y $f_2$ argumentos de entrada (por defecto $10$ y $100$) y $t \in [0, 0.5]$. Además, debe mostrar:
* leyenda,
* título "Dos frecuencias",
* eje x "Tiempo ($t$)"
y usar algún estilo de los disponibles.


```python
def frecuencias(f1=10.0, f2=100.0):
    max_time = 0.5
    times = np.linspace(0, max_time, 10000)
    signal = np.sin(2*np.pi*f1*times) + np.sin(2*np.pi*f2*times)
    
    with plt.style.context('seaborn-dark-palette'):
        plt.plot(signal,label="Señal")
        plt.title("Dos frecuencias")
        plt.xlabel("Tiempo(t)")
        plt.legend()

frecuencias()        
```

**Ejercicio**
Representar las curvas de nivel de esta función:
$$g(x, y) = \cos{x} + \sin^2{y}$$
Para obtener este resultado:

```python
def g(x,y):
    return np.cos(x) + np.sin(y)**2
            
x = np.linspace(-2,3,100)
y = np.linspace(-2,3,100)
z = g(x,y)
with plt.style.context('seaborn-dark-palette'):        
    plt.title("g(x,y) = cos(x) + sin**2")
    plt.xlabel("Eje x")
    plt.ylabel("Eje y")
plt.plot(z)

    
```
