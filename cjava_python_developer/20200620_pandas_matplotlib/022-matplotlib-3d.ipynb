## GRAFICOS EN 3D

Graficamos un mapa de calor

```python
%matplotlib inline
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Nos permite ingresar ejes tridimensionales en nuestra gr√°fica
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot')
```

Creamos figura

```python
fig = plt.figure()
axl = fig.add_subplot(111, projection='3d')
```

Coordenada de barra

```python
x = [1,2,3,4,2,6,7,8,9,10]
y = [3,4,6,7,1,2,5,9,10,5]
z = np.zeros(10)

dx = np.ones(10)
dy = np.ones(10)
dz = [2,3,4,3,4,5,3,7,8,2]
```

Presentacion de los datos

```python
axl.bar3d(x, y, z, dx, dy, dz)
plt.show()
```
