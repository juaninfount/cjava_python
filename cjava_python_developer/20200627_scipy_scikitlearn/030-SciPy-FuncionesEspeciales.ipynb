## FUNCIONES ESPECIALES

Las funciones disponibles en el paquete especial son funciones universales, que siguen a la transmisión y al bucle automático de matrices.

Veamos algunas de las funciones especiales más utilizadas:

Función de raíz cúbica
Funcion exponencial
Función exponencial de error relativo
Función exponencial de suma de registro
Función de Lambert
Función de permutaciones y combinaciones
Función gamma
Ahora comprendamos cada una de estas funciones en breve.

Función de raíz cúbica
La sintaxis de esta función de raíz cúbica es: scipy.special.cbrt (x). Esto buscará la raíz del cubo en cuanto a elementos de x .

Consideremos el siguiente ejemplo.

### Función de raíz cúbica

La sintaxis de esta función de raíz cúbica es: scipy.special.cbrt (x). Esto buscará la raíz del cubo en cuanto a elementos de x .

Consideremos el siguiente ejemplo.

```python
from scipy.special import cbrt
res = cbrt([10, 9, 0.1254, 234])
print(res)
```

### Funcion exponencial

La sintaxis de la función exponencial es - scipy.special.exp10 (x). Esto calculará 10 ** x elemento sabio.

Consideremos el siguiente ejemplo.

```python
from scipy.special import exp10
res = exp10([2, 9])
print(res)
```

### Función exponencial de error relativo

La sintaxis para esta función es: scipy.special.exprel (x). Genera el error relativo exponencial, (exp (x) - 1) / x.

Cuando x está cerca de cero, exp (x) está cerca de 1, por lo que el cálculo numérico de exp (x) - 1 puede sufrir una pérdida catastrófica de precisión. Luego se implementa exprel (x) para evitar la pérdida de precisión, que ocurre cuando x está cerca de cero.

Consideremos el siguiente ejemplo.

```python
from scipy.special import exprel
res = exprel([-0.25, -0.1, 0, 0.1, 0.25])
print(res)
```

### Función exponencial de suma de registro

La sintaxis para esta función es: scipy.special.logsumexp (x). Ayuda a calcular el registro de la suma de exponenciales de los elementos de entrada.

Consideremos el siguiente ejemplo.

```python
from scipy.special import logsumexp
import numpy as np
a = np.arange(10)
res = logsumexp(a)
print(res)
```

### Función de Lambert

La sintaxis para esta función es - scipy.special.lambertw (x). También se llama como la función Lambert W. La función W de Lambert W (z) se define como la función inversa de w * exp (w). En otras palabras, el valor de W (z) es tal que z = W (z) * exp (W (z)) para cualquier número complejo z.

La función Lambert W es una función multivalor con infinitas ramas. Cada rama da una solución separada de la ecuación z = w exp (w). Aquí, las ramas están indexadas por el entero k.

Consideremos el siguiente ejemplo. Aquí, la función Lambert W es la inversa de w exp (w).

```python
from scipy.special import lambertw
w = lambertw(1)
print(w)
print(w * np.exp(w))
```

### Permutaciones y combinaciones

Discutamos las permutaciones y combinaciones por separado para comprenderlas claramente.

Combinaciones : la función de sintaxis para combinaciones es: scipy.special.comb (N, k). Consideremos el siguiente ejemplo:

```python
from scipy.special import comb
res = comb(10, 3, exact = False,repetition=True)
print(res)
```

Nota - Los argumentos de matriz se aceptan solo para el caso exacto = Falso. Si k> N, N <0 o k <0, se devuelve un 0.

Permutaciones : la función de sintaxis para combinaciones es: scipy.special.perm (N, k). Permutaciones de N cosas tomadas k a la vez, es decir, k-permutaciones de N. Esto también se conoce como "permutaciones parciales".

Consideremos el siguiente ejemplo.

```python
from scipy.special import perm
res = perm(10, 3, exact = True)
print(res)
```

## Función gamma

La función gamma a menudo se denomina factorial generalizado ya que z * gamma (z) = gamma (z + 1) y gamma (n + 1) = n !, para un número natural 'n'.

La sintaxis para la función de combinaciones es: scipy.special.gamma (x). Permutaciones de N cosas tomadas k a la vez, es decir, k-permutaciones de N. Esto también se conoce como "permutaciones parciales".

La sintaxis para la función de combinaciones es: scipy.special.gamma (x). Permutaciones de N cosas tomadas k a la vez, es decir, k-permutaciones de N. Esto también se conoce como "permutaciones parciales".

```python
from scipy.special import gamma
res = gamma([0, 0.5, 1, 5])
print(res)
```
